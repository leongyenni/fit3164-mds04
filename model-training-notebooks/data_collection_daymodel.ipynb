```python
import os
from datetime import datetime, timedelta
import matplotlib.pyplot as plt
import pandas as pd
import yfinance as yf

TICKER_PATH='/content/drive/MyDrive/Colab Notebooks/ticker.txt'
DATAS_PATH='/content/drive/MyDrive/Colab Notebooks/datas'
OUTPUT_PATH='/content/drive/MyDrive/Colab Notebooks/outputs'
```

```python
def download_from_yahoo_finance(ticker, start_date, end_date, ticker_path):
    data = yf.download(ticker, interval='1h', start=start_date, end=end_date)
    if not data.empty:
        data.reset_index(inplace=True)
        # data = data[['Datetime', 'Close']]  # Extract 'Date' and 'Close' columns only
        data.to_csv(ticker_path, index=False)
```

```python
def main(start=datetime.now()):
    start = start if not isinstance(start, str) else datetime.strptime(start, '%Y-%m-%d')

    current_market_time = datetime.now() - timedelta(hours=12)
    delta = current_market_time - timedelta(days=1) - start
    dates = [start + timedelta(days=d) for d in range(0, delta.days + 1)]

    with open(TICKER_PATH) as f:
        tickers = [t.rstrip() for t in f if t.rstrip()]

    if not os.path.isdir(DATAS_PATH):
        os.mkdir(DATAS_PATH)

    for ticket in tickers:
        if not os.path.isdir(os.path.join(DATAS_PATH, ticket)):
            os.mkdir(os.path.join(DATAS_PATH, ticket))

        for d in dates:

            if d.weekday() == 5 or d.weekday() == 6:
                continue

            start_date = datetime(d.year, d.month, d.day, 8, 0, 0, 0)
            end_date = d + timedelta(days=1)
            end_date = datetime(end_date.year, end_date.month, end_date.day, 8, 0, 0, 0)

            for t in tickers:
                ticker_path = os.path.join(DATAS_PATH, t)
                ticker_file = os.path.join(ticker_path, '{}.csv'.format(d.date()))

                # Check if file exists and if date is found in the file
                if os.path.isfile(ticker_file):
                    with open(ticker_file, 'r') as f:
                        # Check if the date is in the file
                        if str(d.date()) in f.read():
                            # If date is found, delete the old file
                            os.remove(ticker_file)
                            # Then redownload and save as a new file
                            download_from_yahoo_finance(t, start_date, end_date, ticker_file)
                else:
                    download_from_yahoo_finance(t, start_date, end_date, ticker_file)
```

```python
main("2022-10-01")
```
